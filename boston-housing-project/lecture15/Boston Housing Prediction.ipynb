```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('housing.data',delim_whitespace=True,header=None)
```

```python
df.head()
```

1. CRIM      per capita crime rate by town

2. ZN        proportion of residential land zoned for lots over 25,000 sq.ft.

3. INDUS     proportion of non-retail business acres per town

4. CHAS      Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)

5. NOX       nitric oxides concentration (parts per 10 million)

6. RM        average number of rooms per dwelling

7. AGE       proportion of owner-occupied units built prior to 1940

8. DIS       weighted distances to five Boston employment centres

9. RAD       index of accessibility to radial highways

10. TAX      full-value property-tax rate per $10,000

11. PTRATIO  pupil-teacher ratio by town

12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town

13. LSTAT    % lower status of the population

14. MEDV     Median value of owner-occupied homes in $1000's

```python
col_name=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
```

```python
df.columns=col_name
```

```python
plt.figure(figsize=(18,13))
sns.heatmap(df.corr(),annot=True,cmap='YlGnBu')
plt.show()
```

```python
X = df['RM'].values.reshape(-1,1)
```

```python
y = df['MEDV'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(12,10))
sns.regplot(X,y)
plt.xlabel("Av. No of Rooms per Dwelling")
plt.ylabel("Median val. of owner-occupied homes in $1000's")
plt.grid()
plt.show()
```

```python
sns.jointplot(x='RM', y='MEDV',data=df, kind='reg',height=10)
plt.grid()
plt.show()
```

```python
X = df['LSTAT'].values.reshape(-1,1)
y = df['MEDV'].values
model.fit(X,y)
plt.figure(figsize=(12,10))
sns.regplot(X,y)
plt.xlabel("% of lower status of the population")
plt.ylabel("Median val. of owner-occupied homes in $1000's")
plt.grid()
plt.show()
```

```python
sns.jointplot(x='LSTAT', y='MEDV',data=df, kind='reg',height=10)
plt.grid()
plt.show()
```
